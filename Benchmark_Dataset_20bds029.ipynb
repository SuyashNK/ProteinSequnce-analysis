```python
def create_word2vec_embeddings(data, model_W2V):
    embeddings = []
    sentences = data
    for sentence in sentences:
        vectors = [model_W2V.wv[word] for word in sentence if word in model_W2V.wv]
        if vectors:
            doc_vector = np.mean(vectors, axis=0)
        else:
            doc_vector = np.zeros(model_W2V.vector_size)
        embeddings.append(doc_vector)
    return embeddings
```

```python

def create_fasttext_embeddings(data, model_FT):
    embeddings = []
    sentences = data
    for sentence in sentences:
        vectors = [model_FT.wv[word] for word in sentence if word in model_FT.wv]
        if vectors:
            doc_vector = np.mean(vectors, axis=0)
        else:
            doc_vector = np.zeros(model_FT.vector_size)
        embeddings.append(doc_vector)

    return embeddings
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.models import Sequential
```

```python
def skipxgram_bigrams(input_string,x):
    skip_1_gram_bigrams = []
    for i in range(len(input_string) - x):
        bigram = input_string[i] + input_string[i + x]
        skip_1_gram_bigrams.append(bigram)
    bigram_counts = Counter(skip_1_gram_bigrams)
    vector_length = 400
    frequency_vector = [0] * vector_length
    for i in range(vector_length):
        bigram = chr(65 + (i // 20)) + chr(65 + (i % 20))
        frequency_vector[i] = bigram_counts.get(bigram, 0)

    return frequency_vector
```

```python
final_df['skip5gram'] = final_df['Sequence'].apply(lambda x: skipxgram_bigrams(x,5))
```

## Defining the Model

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

tf.keras.backend.set_floatx('float64')

# Adjusting input shape for the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(2048, activation='relu', input_shape=(400,)), # specifying the input shape  # specifying the input shape
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(5, activation='sigmoid'),  # 5 neurons for 5 labels
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy'])

# Assuming final_df has columns for each label e.g., 'envelope', 'lumen', ...
X = final_df['skip5gram']
y = final_df[['envelope', 'lumen', 'plastoglobule', 'stroma', 'thylakoid_membrane']]  # all labels

from sklearn.model_selection import train_test_split
X_train_list, X_test_list, Y_train, Y_test = train_test_split(X, y, test_size=0.2)

# Convert lists of arrays to tensors and then reshape
X_train = tf.reshape(tf.stack([tf.convert_to_tensor(arr, dtype=tf.float32) for arr in X_train_list]),
                     (len(X_train_list), 400))
X_test = tf.reshape(tf.stack([tf.convert_to_tensor(arr, dtype=tf.float32) for arr in X_test_list]),
                    (len(X_test_list), 400))

model.fit(X_train, Y_train, epochs=10)


```

### Results

```python
test_loss, test_acc = model.evaluate(X_test,  Y_test, verbose=2)

print('\nTest accuracy:', test_acc)
```

```python
def compute_strict_accuracy(y_true, y_pred):
    """Compute the strict accuracy for multi-label classification."""
    y_pred_binary = (y_pred > 0.5).astype(int)
    correct_predictions = (y_true == y_pred_binary).all(axis=1)
    strict_accuracy = correct_predictions.mean()
    return strict_accuracy

# Create a list to store strict accuracies
strict_accuracies = []

# Iterate over the dataset and compute strict accuracy for each sequence
for index, row in final_df.iterrows():
    x_input = tf.reshape(tf.convert_to_tensor(row['skip5gram'], dtype=tf.float32), (1, 400))
    y_true = row[['envelope', 'lumen', 'plastoglobule', 'stroma', 'thylakoid_membrane']].values.reshape(1, -1)

    y_pred = model.predict(x_input)
    accuracy = compute_strict_accuracy(y_true, y_pred)
    strict_accuracies.append(accuracy)

# Create a new DataFrame
results_df = pd.DataFrame({
    'PDBid': final_df['PDBid'],
    'Sequence': final_df['Sequence'],
    'Strict Accuracy': strict_accuracies
})

# Save to CSV
results_df.to_csv('Strict_accuracies_Benchmark.csv', index=False)
```

```python
def compute_relaxed_accuracy(y_true, y_pred):
    """Compute the relaxed accuracy for multi-label classification."""
    y_pred_binary = (y_pred > 0.5).astype(int)

    # Calculate correctly predicted labels for each data point
    correct_predictions = (y_true == y_pred_binary).sum(axis=1)

    # Divide by total number of labels to get relaxed accuracy for each data point
    relaxed_accuracy = correct_predictions / y_true.shape[1]
    return relaxed_accuracy

# Create a list to store relaxed accuracies
relaxed_accuracies = []

# Iterate over the dataset and compute relaxed accuracy for each sequence
for index, row in final_df.iterrows():
    x_input = tf.reshape(tf.convert_to_tensor(row['skip5gram'], dtype=tf.float32), (1, 400))
    y_true = row[['envelope', 'lumen', 'plastoglobule', 'stroma', 'thylakoid_membrane']].values.reshape(1, -1)

    y_pred = model.predict(x_input)
    accuracy = compute_relaxed_accuracy(y_true, y_pred)
    relaxed_accuracies.append(accuracy[0])  # Get the scalar value

# Create a new DataFrame
results_df = pd.DataFrame({
    'PDBid': final_df['PDBid'],
    'Sequence': final_df['Sequence'],
    'Relaxed Accuracy': relaxed_accuracies
})

# Save to CSV
#results_df.to_csv('sequence_relaxed_accuracies.csv', index=False)
```
